## a) El Modelo Ingreso-Gasto: la Curva IS
 
### - Utilizando las ecuaciones derive paso a paso la curva IS matemáticamente (a partir de la condición de equilibrio Y = DA).

DA es Demanda Agregada, la cual podemos expresar de la siguiente manera:

$$Y = DA = C + I + G + T + X - M$$

en donde podemos calcular cada variable:

$$C = Co +b Yd$$

$$I = Io - hr$$

$$G = Go$$

$$X = Xo$$

$$M = mYd$$

$$Tributacion (T) = tY$$

Sin embargo, para hallar el equilibrio que existe entre el ahorro y la inversion, debemos quitar la tributacion (T), por lo que, de la primera ecuacion, restaremos la T, quedando de esta forma: 

$$Y  - T = C + I + G + X - M$$
$$Y^d= C + I - T + G + X - M$$

Que se puede escribir de la siguiente forma:
$$(Y^d-C)+(T-G)+(M-X)=I$$

Por otro lado, tenemos al ahorro, el cual esta compuesto por el ahorro privado (Sp), ahorro del gobierno (Sg) y ahorro externo (Se). Entonces si el Ahorro (S), es igual a la inversion (I):

$$S = Sp + Sg + Se$$

Desglosamos esta ecuacion para obtener la curva IS: 

$$I = S_p + S_g + S_e$$

$$I(r) = S(Y)$$

$$I_o - hr = Sp + Sg + Se$$

$$I_o - hr = (Y^d - C_o - bYd^) + (T - G_o) + (mY^d - X_o)$$

$$I_o - hr = [1 - (b - m)(1 - t)]Y - (C_o - G_o + X_o)$$


Finalmente, para que este en funcion del ingreso:

$$hr=(C_o - G_o + X_o + I_o)-(1 - (b - m)(1 - t))Y$$
$$r=\frac{1}{h}(C_o - G_o + X_o + I_o)-\frac{1 - (b - m)(1 - t)}{h}y$$

Que puede simplicarse en 

$$r=\frac{B_0}{h}-\frac{B_1}{h}$$

Donde $B_0=C_o - G_o + X_o + I_o$ es el intercepto y $B_1=(1 - (b - m)(1 - t)$

### Utilizando la función de equilibrio IS donde r está en función de Y, encuentre △r/△Y que representa la pendiente y qué signo tiene?

$$\frac{(1-t)(b-m)-1}{h}$$ 

La pendiente es negativa: $\frac{1+(b-m)(1-t)}{h}$

### Lea la sección 4.4 del material de enseñanza y explique cómo se deriva la curva IS a partir del equilibrio Y=DA y grafique

Ecuacion del ingreso de equilibrio a corto plazo:
$$Y+\frac{1}{1-(b-m)(1-t)}(C_o - G_o + X_o + I_o-hr)$$


Expresada en funcion a la tasa de interes:
$$r=\frac{1}{h}(C_o - G_o + X_o + I_o)-\frac{1 - (b - m)(1 - t)}{h}$$

que puede simplificarse:

$$r=\frac{B_0}{h}-\frac{B_1}{h}Y$$

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Segunda curva de ingreso de equilibrio

# Definir cualquier parámetro autónomo
Go = 35

# Generar la ecuación con el nuevo parámetro
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go =70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
```
