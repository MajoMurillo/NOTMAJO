## Pregunta 1
### Derive y explique la función de demanda de Consumo
Cuando hablamos de demanda de Consumo, nos referimos al gasto que hacen las familias y la distribucion del dinero en donde:

- T = tributacion total, es decir, todo lo que el pais puede recaudar de nuestros impuestos
- t = tasa impositiva, la cual varia de 0 (no paga impuestos) a 1 (paga mas impuestos de lo debido)

Gracias a estas dos variables, tendremos nuestra primera formula la cual sera

$$T = t * Y$$

Para hallar la funcion del consumo, tenemos que tomar en cuenta el ingreso disponible (Yd), ya que es el dinero que quedara a las familias sin los impuestos, cuya formula es:
$$Yd=Y−T$$

Entonces $Yd = (1 − t) ∗ Y$

Sabiendo el ingreso disponible, podemos hallar la funcion de consumo, la cual seria

$$C=Co⁡+b ∗ Yd$$

donde Co = consumo fijo y b = gastos variables

```python
# Parámetros

Yd_size = 70 

Co = 10
b = 0.5 # b > m

Yd = np.arange(Yd_size)

def C_K(Co, b, Yd):
    C_K = Co + (b*Yd)
    return C_K

C_IS_K = C_K(Co, b, Yd)
```

```python
# Gráfico

y_max = np.max(C_IS_K)
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(Yd, C_IS_K, label = "C=Co+bYd", color = "#D2691E") 

plt.text(40, 43, '$PMgC = {∆C}/{∆Y^d} = b = 0.5$', fontsize = 12, color = 'black')

ax.set(title="Función de demanda de Consumo", xlabel= r'Yd', ylabel= r'C')
ax.legend() 

custom_xlim = (0, 80)
custom_ylim = (0, 80)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

## Pregunta 2
### Derive y explique la función de Inversion (I)
La inversion son los gastos que aumentan el stock de la economia, donde:

- Io = inversión autónoma

- r = tasa de interés

- h: sensibilidad que tienen los inversionistas ante la subida o bajada de la tasa de interes

Tomemos en cuenta que si la inversion autonoma sube, entonces (la sensibilidad * tasa de interes) baja

$$I=Io−hr$$

```python
# Parámetros

r_size = 100 


Io = 80
h = 0.7
r = 0.9

r = np.arange(r_size)

def I_K(Io, h, r):
    I_K = Io - (h)*(r)
    return I_K

I_IS_K = I_K(Io, h, r)
```

```python
# Gráfico

y_max = np.max(I_IS_K)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(r, I_IS_K, label = "I=Io-hr", color = "#006400")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(90, 20, '$I=I_0-hr$', fontsize = 12, color = 'black')

ax.set(title="Función de demanda de Inversión", xlabel= r'r', ylabel= r'I')
ax.legend() 

custom_xlim = (0, 120)
custom_ylim = (-10, 120)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

## Pregunta 4
### Encuentre el nivel de Ingreso de equilibrio (Ye) 
Tomemos en cuenta que el ingreso de equilibrio es exactamente igual a la demanda agregada de equilibrio ya que estan sobre un grafico de 45 grados, es decir DA = Ye

Donde:
- Consumo = $Co + bYd$
- Inversion= $I = Io - hr$
- Gobierno= $G = Go$
- Tributacion= $T= tY$
- Exportaciones $X=X0$
- Importaciones= $M=mYd$
- Demanda agregada= $DA=C+I+G+X-M$

Entonces:
$$DA = C + I + G + X - M$$

$$DA = Co + b(1-t)Y + Io - hr + Go + Xo - m(1-t)Y$$

$$DA = Co + Io - hr + Go + Xo + b(1-t)Y - m(1-t)Y$$

$$DA= Co + Io - hr + Go + Xo + (b-m)(1-t)Y$$


Reducimos $Co + Io - hr + Go + Xo$ a $α0$ el cual es el intercepto y $(b-m)(1-t)$ a $α1$ que es la pendiente de la funcion, y nos quedaria de esta manera:

$$DA= α_0 + α_1Y$$


Considerando que $Y=DA$
$$Y = α_0 + α_1Y$$

$$Y^e=C_0+bY^d+I_0-hr+G_0+X_0-mY^d$$

$$Y^e= \frac{1}{1−(b−m)(1−t)}(Co+Io+Go+Xo−hr)$$

Donde $\frac{1}{1-(b-m)(1-t)}$ es el multiplicador keynesiano

```python
Y_size = 100
Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9
Y = np.arange(Y_size)
```

```python
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
a = 2.5

def L_45(a, Y):
 L_45 = a*Y
 return L_45
L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#8FBC8B") #Demanda agregada
ax.plot(Y, L_45, color = "#FF4500") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5, ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=145, xmin= 0, xmax= 0.8, linestyle = ":", color = "grey")

# Texto agregado
plt.text(85, 192, '$DA=α_o + α_1Y$', fontsize = 11.5, color = 'black')
plt.text(82, 170, '$α_1=(b-m)(1-t) $', fontsize = 11.5, color = 'black')
plt.text(70, 110, '$α_o=(Co+Io+Go+Xo-hr) $', fontsize = 11.5, color = 'black')
plt.text(70, 130, '$↑$', fontsize = 18, color = 'black')
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

# línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda
plt.show()
```

## Pregunta 5
### Analice una Política Fiscal Expansiva con aumento del Gasto del Gobierno
**-INTUICION**
$$↑Go →↑DA →DA>Y →↑Y$$

-**MATEMATICAMENTE**
$$Y^e= \frac{1}{1−(b−m)(1−t)}(Co+Io+Go+Xo−hr)$$

Considerando el multiplicador keynesiano:

$$Y=k(Co+Io+Go+Xo−hr)$$

$$ΔY=Δk(ΔCo+ΔIo+ΔGo+ΔXo−Δhr)$$


Si no ha habido cambios en Co, Io, Go, Xo, h ni r entonces:

$$ΔCo=ΔIo=ΔGo=ΔXo=Δh=Δr=0$$

$$ΔY=k(ΔGo)$$

Sabiendo que ambos >0, entonces:

$$ΔY=(+)(+)$$

$$ΔY>0$$

Donde $\frac{1}{-(1-t)(b-m)+1}$ es el diferencial:

$$\frac{ΔY}{ΔGo}=\frac{1}{1-(b-m)(1-t)}$$

Sabiendo que Go y k son mayores que 0:

$$\frac{ΔY}{(+)}=(+)$$

$$ΔY=(+)(+)$$

$$ΔY>0$$

```python
Y_size = 100
Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
Y = np.arange(Y_size)

 # Ecuación
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K
DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio
# Se define el parámetro cambiado
Go = 90
# Generar la ecuación con el nuevo parámetro
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K
DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# 45°
a = 2.5
def L_45(a, Y):
 L_45 = a*Y
 return L_45
L_45 = L_45(a, Y)
```

### ¿Cuál es el papel que juega el parametro  en el multiplicador keynesiano?
En primer lugar el multiplicador keynesiano es un cambio que ocurre entre los componentes de la demanda agregada (DA), lo cual acaba formando un nuevo producto de equilibrio.
El papel del parametro es que "el tamano del multiplicador keynesiano depende de la magnitud de los parametros" (Jimenez, F. "Apuntes de Macroeconomia")

```python
# líneas punteadas autómaticas

    # definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [DA_Cont[0], Y[0]] # DA, coordenada inicio
B = [DA_Cont[-1], Y[-1]] # DA, coordenada fin

C = [L_45[0], Y[0]] # L_45, coordenada inicio
D = [L_45[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec # (y,x)
```
