### A partir del siguiente sistema de ecuaciones que representan el modelo IS-LM
#### - Encuentre las ecuaciones de Ingreso ($Y^e$) y tasa de interes de equilibrio ($r^e$) (Escriba paso a paso la derivacion de estas ecuaciones)
Curva IS:
$$r = \frac{B_0}{h} - \frac{B_1}{h}Y$$

donde:
- $B_0 = C_0+I_0+G_0+X_0$

- $B_1=1-(b-m)(1-t)$

Curva LM:

$$r=-\frac{1}{j}\frac{M_0}{P}+\frac{k}{j}Y$$

Entonces para encontrar el **equilibrio del Ingresos $(Y_e)$**, se tienen que igualar ambar ecuaciones y despejar Y

$$\frac{B_0}{h}-\frac{B_1}{h}Y=-\frac{1}{j}\frac{M_0}{P}+\frac{k}{j}Y$$

$$j(B_0)-j(B_1)Y=h(k)Y-h\frac{M_0}{P}$$

$$j(B_0)+h\frac{M_0}{P}=(kh+jB_1)Y$$

$$Y^e=\frac{jB_0}{kh+jB_1}+\frac{h}{kh+jB_1}\frac{M_0}{P}$$


Mientras que para hallar **la tasa de interes de equilibrio**, se tiene que despejar Y de ambas curvas

Curva Is:

$$r = \frac{B_0}{h} - \frac{B_1}{h}Y$$

$$rh=B_0-B_1Y$$

$$B_1Y=B_0-rh$$

$$Y=\frac{B_0-rh}{B_1}$$

Curva LM:

$$r=-\frac{1}{j}\frac{M_0}{P}+\frac{k}{j}$$

$$rj+\frac{M_0}{P}=kY$$

$$\frac{rj}{k}+\frac{1}{k}\frac{M_0}{P}=Y$$

Se igualan y se despeja r:

$$\frac{B_0-rh}{B_1}=\frac{rj}{k}+\frac{1}{k}\frac{M_0}{P}$$

$$kB_0-k(rh)=B_1(rj)+B_1\frac{M_0}{P}$$

$$kB_0-B_1\frac{M_0}{P}=B_1(rj)+k(rh)$$

$$kB_0-B_1\frac{M_0}{P}=r(B_1j+kh)$$

$$r_e=\frac{kB_0}{kh+jB1}-\frac{B_1}{kh+jB_1}\frac{M_0}{P}$$

#### - Grafique el equilibrio simultáneo en los mercados de bienes y de dinero.

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 40
Io = 45
Go = 50
Xo = 2
h = 0.8
b = 0.4
m = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo)/h - ( ( 1-(b-m)*(1-t) ) / h)*Y  
    return r_IS

r_is = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 180            
P  = 25               

Y = np.arange(Y_size)

# Ecuación

def r_LM(k, j, Ms, P, Y):
    r_LM = - (1/j)*(Ms/P) + (k/j)*Y
    return r_LM

r_lm = r_LM( k, j, Ms, P, Y)
```

```python
# Gráfico del modelo IS-LM

# Dimensiones del gráfico
y_max = np.max(r_lm)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
# Curva IS
ax.plot(Y, r_is, label = "IS", color = "MediumSeaGreen") #IS
# Curva LM
ax.plot(Y, r_lm, label="LM", color = "Crimson")  #LM

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto y figuras agregadas
# Graficar la linea horizontal - r
plt.axvline(x=51.5,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
# Grafica la linea vertical - Y
plt.axhline(y=93, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")

# Plotear los textos 
plt.text(49,100, '$E_0$', fontsize = 14, color = 'black')
plt.text(0,100, '$r_0$', fontsize = 12, color = 'black')
plt.text(53,-10, '$Y_0$', fontsize = 12, color = 'black')

# Título, ejes y leyenda
ax.set(title="Grafico simultaneo en los mercados de dinero y de bienes", xlabel= r'Y', ylabel= r'r')
ax.legend()

plt.show()
```

$$r=-\frac{1}{j}\frac{M_0}{P}+\frac{k}{j}Y$$

Podemos igualar, sustituir o reducir ambas ecuaciones para encontrar el nivel de Ingresos equilibrio y la tasa de interés de equilibrio

$$\frac{B_0}{h}-\frac{B_1}{h}Y=-\frac{1}{j}\frac{M_0}{P}+\frac{k}{j}Y$$

Ingreso de equilibrio:

$$Y^e=\frac{jB_0}{kh+jB_1}+\frac{h}{kh+jB_1}\frac{M_0}{P}$$

Tasa de interes de equilibrio:

$$r_e=\frac{kB_0}{kh+jB1}-\frac{B_1}{kh+jB_1}\frac{M_0}{P}$$

Estas dos ecuaciones representan el modelo IS-LM

### 1.2 Grafique el equilibrio simultáneo en los mercados de bienes y de dinero.

```python
#--------------------------------------------------
    # Curva IS

# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 50
h = 0.8
b = 0.5
t = 0.8

Y = np.arange(Y_size)


# Ecuación 
def r_IS_2(b, t, Co, Io, Go, h, Y):
    r_IS_2 = (Co + Io + Go - Y * (1-b*(1-t)))/h
    return r_IS_2

r_2 = r_IS_2(b, t, Co, Io, Go, h, Y)


#--------------------------------------------------
    # Curva LM 

# Parámetros

Y_size = 100

k = 2
j = 1                
Ms = 200             
P  = 20
π = 4

Y = np.arange(Y_size)

# Ecuación

def i_LM_2(k, j, Ms, P, Y, π):
    i_LM_2 = (-Ms/P)/j + k/j*Y - π
    return i_LM_2

i_2 = i_LM_2( k, j, Ms, P, Y, π)
```
